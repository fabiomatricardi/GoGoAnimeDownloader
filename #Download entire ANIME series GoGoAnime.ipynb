````python
import requests
 
# Using the example query "demon", and looking at the 2nd page of results.
url = f"https://api.consumet.org/anime/gogoanime/{myidx}"
response = requests.get(url, params={"page": 1})
search = response.json()
 
print(search)
````

````python
import requests
 
# Using the example ID of "hundred".
url = f"https://api.consumet.org/anime/gogoanime/info/{myidx}"
response = requests.get(url)
info = response.json()
title =info['title']
eps = info['totalEpisodes']
idname = info['id']
print(info['title'])
print("---------------")
print(info['description'])
print("Total Num of Episodes: "+str(info['totalEpisodes']))
````

````python
%%capture
#Download silent
"""
Using the example episode ID taken from info[id],
explicitly defining default server for demostrative purposes.
"""
for ep in range(1,eps+1):
  url = f"https://api.consumet.org/anime/gogoanime/watch/{idname}-episode-{ep}"
  response = requests.get(url, params={"server": "vidstreaming"})
  data = response.json()
  link = data['sources'][0]['url']
  filename=f"/content/{title}_ep_{ep}.mp4"
  !~/.local/bin/downloadm3u8 -o "$filename" "$link"
````

````python
#### PREPARE FOR SERVER RUNNING  ##################################
# START IN SUBPROCESS YOUR WEBSERVER
#===================================
from IPython.core.magic import register_line_magic
import IPython
import subprocess

@register_line_magic
def run_local_server(line):
    handle = IPython.display.display(
            IPython.display.Pretty("Launching my server..."),
            display_id=True,
    )
    subprocess.Popen(['python', '-m', 'http.server'])
    shell = """
        (async () => {
            const url = new URL(await google.colab.kernel.proxyPort(8000, {'cache': true}));
            const iframe = document.createElement('iframe');
            iframe.src = url;
            iframe.setAttribute('width', '100%');
            iframe.setAttribute('height', '400');
            iframe.setAttribute('frameborder', 0);
            document.body.appendChild(iframe);
        })();
    """
    script = IPython.display.Javascript(shell)
    handle.update(script)
````
